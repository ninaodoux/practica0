```python
path ="../data/titanic1.csv"
titanic_df = pd.read_csv(path, sep=';', low_memory=False)
print(titanic_df)
```

```python
#tenemos 887 obs y 8 variables
#Ahora importamos librerias 

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report,ConfusionMatrixDisplay
from sklearn import tree 
```

```python
import matplotlib.pyplot as mplt
```

```python
#para ver las columnas 
print(titanic_df.columns)
```

```python
#preprocesar los datos :

#verificar el tipo con

print(titanic_df.info())
```

```python
#convertimos las numericas

titanic_df['Sex'] = titanic_df['Sex'].map({'male':0, 'female':1})

#llenar los valores faltantes NaN cuando hay 
titanic_df.isna().sum()

#no hace falta 


```

```python
#separar variables independientes y dependientes 
#variables independientes 

x= titanic_df[['Pclass', 'Sex', 'Age', 'Siblings/Spouses Aboard', 'Parents/Children Aboard', 'Fare']]

#las dependientes: porque establecemos que vivir o sobrevivir depende de las variables x

y= titanic_df['Survived']
```

```python
#Dividir los datos en TRAIN y TEST 

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
#construccion del arbol de classification 

#arbol de decision 
clf = DecisionTreeClassifier()

#Entrenar el classif

clf.fit(x_train, y_train)

#predecir en el cojunto de TEST

y_pred= clf.predict(x_test)

```

```python
#evaluar el rendimiento del modelo 

from sklearn import metrics

print("Accuracy:", metrics.accuracy_score(y_test,y_pred))
print("Confusion Matrix:\n", metrics.classification_report(y_test,y_pred))
print("Classification report:\n", metrics.classification_report(y_test,y_pred))
```

```python
#visualizacion del arbol 

import matplotlib.pyplot as mplt 
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(clf, feature_names=['Pclass', 'Sex', 'Age', 'Siblings/Spouses Aboard', 'Parents/Children Aboard', 'Fare'], class_names=['Not Survived', 'Survived'])
plt.show()
```
